```python
embedding_matrix = torch.tensor([
    [0.1, 0.2, 0.3],
    [1.1, 1.2, 1.3],
    [2.1, 2.2, 2.3],
    [3.1, 3.2, 3.3],
    [4.1, 4.2, 4.3]
])

token_ids = torch.tensor([2,4,1])
output = embedding_matrix[token_ids]

print("Embedding Matrix: ")
print(embedding_matrix)

print("output: ")
print(output)

print("output Shape: ")
print(output.shape)
```
