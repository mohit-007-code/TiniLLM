```python
sentence = ""

words = ["I", "Love", "Python"]
for word in words:
    sentence += word + " "
    print(sentence)
```

```python
sentence = "I Love Python"
words = sentence.split()

for i in range(1, len(words)):
    context = words[:i]
    target = words[i]

    print(context)
    print(target)
    print()
```
