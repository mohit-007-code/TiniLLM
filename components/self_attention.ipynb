```python
import torch
import torch.nn as nn
```

```python
class SelfAttention(nn.Module):
    def __init__(self, d_model):
        super().__init__()

        #weight matrix for query
        self.W_q = nn.Parameter(torch.randn(d_model, d_model))

        #weight matrix for key
        self.W_k = nn.Parameter(torch.randn(d_model, d_model))

        #weight matrix for value
        self.W_v = nn.Parameter(torch.randn(d_model, d_model))

    def forward(self, x):
        Q = x @ self.W_q
        K = x @ self.W_k
        V = x @ self.W_v

        return Q, K, V
```

```python
batch_size = 2
sequence_lenght = 6
d_model = 4

x = torch.rand(batch_size, sequence_lenght, d_model)
attention = SelfAttention(d_model)

Q, K, V = attention(x)

print("Input Shape :", x.shape)
print("Query Shape :", Q.shape)
print("Key Shape   :", K.shape)
print("Value Shape :", V.shape)
```
