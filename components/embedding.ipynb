```python
import torch
import math
import torch.nn as nn
```

```python
class Embedding(nn.Module):
    def __init__(self, vocab_size, embedding_dim):
        super().__init__()

        self.embedding_matrix = nn.Parameter(
            torch.randn(vocab_size, embedding_dim)
        )

    def forward(self, token_ids):
        return self.embedding_matrix[token_ids]
```

```python
embedding = Embedding(
    vocab_size=50000,
    embedding_dim=512
)

token_ids = torch.tensor([[2,4,1], [1,2,3]])
output = embedding(token_ids)

print('Input Token IDs: ')
print(token_ids)

print("Output Embeddings: ")
print(output)

print("Output Shape:")
print(output.shape)
```

```python
class PositionalEncoding(nn.Module):
    def __init__(self, d_model, max_seq_length):
        super().__init__()

        pe = torch.zeros(max_seq_length, d_model)

        position = torch.arange(max_seq_length).unsqueeze(1).float()

        div_term = torch.exp(
            torch.arange(0, d_model, 2).float()* (-math.log(10000.0)/d_model)
        )

        pe[:, 1::2] = torch.sin(position*div_term)
        pe[:, 1::2] = torch.cos(position*div_term)

        pe = pe.unsqueeze(0)

        self.register_buffer("pe", pe)

    def forward(self, x):
        seq_lenght = x.size(1)
        return x + self.pe[:, :seq_lenght]
```

```python
d_model = 8
max_seq_length = 10

pe = PositionalEncoding(
    d_model=d_model,
    max_seq_length=max_seq_length
)

print(pe.pe.shape)
```

```python
batch_size = 2
seq_length = 5
d_model = 8

embeddings = torch.randn(batch_size, seq_length, d_model)

output = pe(embeddings)

print("Input Shape :", embeddings.shape)
print("Output Shape:", output.shape)
```
