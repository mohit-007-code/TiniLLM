```python
class SimpleTokenizer:
    def tokenizer(self, text):
        return text.lower().split()
```

```python
tokenizer = SimpleTokenizer()
text = "Hello my name is mohit"
token = tokenizer.tokenizer(text)
print(token)
```

```python
class Vocabulary:
    def __init__(self):
        self.token_to_id = {
            "<PAD>": 0,
            "<UNK>" : 1,
            "<BOS>" : 2,
            "<EOS>" : 3
        }

        self.id_to_token = {
            0: "<PAD>",
            1: "<UNK>",
            2: "<BOS>",
            3: "<EOS>",
        }

    def build(self, tokenized_sentences):
        next_id = len(self.token_to_id)

        for sentence in tokenized_sentences:
            for token in sentence:
                if token not in self.token_to_id:
                    
                    self.token_to_id[token] = next_id
                    self.id_to_token[next_id] = token
                    next_id += 1

    def encode(self, tokens):
        ids = [self.token_to_id["<BOS>"]]

        for token in tokens:
            ids.append(self.token_to_id.get(
                token, self.token_to_id["<UNK>"]
            ))

        ids.append(self.token_to_id["<EOS>"])
        return ids

    def decode(self, ids):
        tokens = []
        for idx in ids:
            tokens.append(self.id_to_token.get(
                idx, "<UNK>"
            ))
        return tokens
```

```python
def main():
    sentences = [
        "I love AI",
        "I love Python",
        "Transformers are amazing"
    ]

    tokenizer = SimpleTokenizer()
    tokenized_sentences = []

    for sentence in sentences:
        tokens = tokenizer.tokenizer(sentence)
        tokenized_sentences.append(tokens)

    print("=="*50)
    print("Tokenized Sentences")
    print("=="*50)

    for tokens in tokenized_sentences:
        print(tokens)


    vocab = Vocabulary()
    vocab.build(tokenized_sentences)

    print("\n" + "=" * 50)
    print("Vocabulary")
    print("=" * 50)

    for token, idx in vocab.token_to_id.items():
        print(f"{token:15} --> {idx}")

    sample_sentence = "I Love Ai"
    sample_tokens = tokenizer.tokenizer(sample_sentence)
    encode = vocab.encode(sample_sentence)

    print("\n" + "=" * 50)
    print("Encoded Sentence")
    print("=" * 50)

    print(sample_tokens)
    print(encode)

    decoded = vocab.decode(encode)

    print("\n" + "=" * 50)
    print("Decoded Sentence")
    print("=" * 50)

    print(decoded)
```

```python
main()
```
